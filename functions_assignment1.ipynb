**6. Explain the concept of generators in Python and how they are defined.**

ans-Generators are a simple way to create iterators using functions and the yield keyword is used to define them instead of return.

for eg:-

```python
def square(n):
  for i in range(n):
    yield(i*i)
print(square(5))
```

```python
for i in square(5):
  print(i)
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

ans-**map function:-** The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.

for eg:-



```python
a=[1,2,6,4]
def func(x):
  if x%2==0:
    return 'the number {} is even'.format(x)
  else:
    return 'the number {} is odd'.format(x)
list(map(func,a))
```

**reduce function:-** The reduce function from the functools module applies a given function cumulatively to the items of a sequence, from left to right, to reduce the sequence to a single value.

for eg:-

```python
from functools import reduce
a=[1,2,3,4]
def func(x,y):
  return x+y
reduce(func,a)
```

**filter function:-** It is used to filter elements from an iterable based on some conditions.

for eg:-

```python
b=['pw', 'skills' , 'data' , 'science']
c= list(filter(lambda x: len(x)>=6, b))
print(c)
```

each number.**


```python
def integers(num):
  a=[]
  for i in num:
    b=i**2
    a.append(b)
  return a
print(integers([3,6,9,12]))
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**


```python
def prime(num):
  if num>1:
    for n in range(2,num):
      if num%n==0:
        print(num,'is not a prime no')
        break
    else:
      print(num,'is a prime no')
  else:
    print(num,'is not a prime no')
print(prime(29))
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**



```python
def fib(n):
  a=0
  b=1
  for i in range(n):
       yield a
       a,b=b,a+b
print(fib(13))
for b in fib(13):
  print(b)
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def exp(exponent):
  for i in range(1, exponent+1):
    yield 2**i
for j in exp(10):
  print(j)
```
